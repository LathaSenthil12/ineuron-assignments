```python
import datetime
df_ap['Date']=pd.to_datetime(df_ap['Date'])
df_lr['Date']=pd.to_datetime(df_lr['Date'])

df_lr['Login Time']=pd.to_datetime(df_lr['Login Time'], format="%H:%M:%S %p").apply(pd.Timestamp)
df_lr['Logout Time']=pd.to_datetime(df_lr['Logout Time'],format='%I:%M:%S').apply(pd.Timestamp)
df_lr['Duration']=pd.to_datetime(df_lr['Duration'],format='%H:%M:%S').apply(pd.Timestamp)
df_ap['Average Response Time']=pd.to_datetime(df_ap['Average Response Time'],format='%H:%M:%S').dt.time
df_ap['Average Resolution Time']=pd.to_datetime(df_ap['Average Resolution Time'],format='%H:%M:%S').dt.time
```
